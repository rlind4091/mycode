```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
# Define coordinates of where to start our map
# 
# hershey = [40.284, -76.649]
# disney = [28.385, -81.563]
# boulder = [40.015, -105.2705]
# kansascity = [39.099, -94.578]

# Define the coordinates we want our markers to be at
polynesian_hotel = [28.404, -81.585]     # famous hotel
epcot_park = [28.374, -81.549]   # family theme park
chocolate_world = [28.425, -81.462]  # local comic shop
start_location = [28.385, -81.563]

# Create the map
polygon_map = folium.Map(location = start_location,  # Latitude and Longitude of Map 
                    zoom_start = 12)            # Initial zoom level for the map (default is 10)
# Add markers to the map
# popup --> label for the Marker; click on the pins on the map!
folium.RegularPolygonMarker(polynesian_hotel, popup = 'Polynesian', fill_color = 'red',
                            number_of_sides = 5, radius = 7).add_to(polygon_map)
folium.RegularPolygonMarker(epcot_park, popup = 'Epcot Park', fill_color = 'yelllow',
                            number_of_sides = 3, radius = 22).add_to(polygon_map)
folium.RegularPolygonMarker(chocolate_world, popup = 'Chocolate World', fill_color = 'green',
                           number_of_sides = 8, radius = 13).add_to(polygon_map)

# change the icon on a marker (location of Hershey Factory)
# see more icons here: https://getbootstrap.com/docs/3.3/components/
folium.Marker([28.385, -81.563], popup = 'Disney', icon=folium.Icon(color="red", icon="info-sign")).add_to(polygon_map)

#Display the map
polygon_map

```

```python
#Adding a tileset to our map
map_with_tiles = folium.Map(location = start_location, tiles = 'Stamen Toner')
map_with_tiles
```

```python
circle_map = folium.Map(location=[28.385, -81.563], tiles="Stamen Toner", zoom_start=13)

folium.Circle(
    radius=100,
    location=[28.385, -81.563],
    popup="Hershey Park",
    color="crimson",
    fill=False,
).add_to(circle_map)

folium.CircleMarker(
    location=[28.385, -81.545],
    radius=50,
    popup="Palmyra",
    color="#3186cc",
    fill = False
    #fill=True,
    #fill_color="#3186cc"
).add_to(circle_map)

circle_map
```
